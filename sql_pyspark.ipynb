```python
from pyspark.sql import functions as F, types as T

row_customers = [
    (1, "John", "Doe", 30, "USA"),
    (2, "Jane", "Smith", 25, "Canada"),
    (3, "Alice", "Johnson", 35, "Mexico"),
    (4, "Bob", "Williams", 40, "Brazil"),
    (None, "Charlie", "Brown", 28, "Argentina"),
]
```

```python
row_orders = [
    (101, 1, 120.0, "USA"), 
    (102, 2, 150.0, "Canada"), 
    (103, 3, 80.0, "Mexico"), 
    (104, 4, 200.0, "Brazil"), 
    (105, None, 90.0, "Argentina"), 
    (106, 1, 180.0, "USA"), 
    (107, 2, 220.0, "Canada"), 
    (108, 3, 100.0, "Mexico"), 
    (109, 4, 250.0, "Brazil"), 
    (110, None, 110.0, "Argentina"),
]
```

```python
schema_customers = T.StructType([
    T.StructField("customer_id", T.IntegerType(), True),
    T.StructField("first_name", T.StringType(), True),
    T.StructField("last_name", T.StringType(), True),
    T.StructField("age", T.IntegerType(), True),
    T.StructField("country", T.StringType(), True),
])
schema_orders = T.StructType([
    T.StructField("order_id", T.IntegerType(), True),
    T.StructField("customer_id", T.IntegerType(), True),
    T.StructField("amount", T.DoubleType(), True),
    T.StructField("country", T.StringType(), True),
])
```

```python
df_customers = spark.createDataFrame(row_customers, schema_customers)
df_orders = spark.createDataFrame(row_orders, schema_orders)
```

```python
df_customers.display()
```

```python
df_orders.display()
```

```python
df_joined = df_orders.join(df_customers, on="customer_id", how="inner")
df_joined.display()
```

```python
o, c = df_orders.alias('o'), df_customers.alias('c')
```

```python
new_view = (
    o.join(c, on=(c.customer_id == o.customer_id), how='inner', )
    .select(c.first_name, c.last_name, o.order_id, o.amount, o.country, c.age,)
)
```

```python
new_view.display()
```
